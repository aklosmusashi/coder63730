```python
import requests
import pandas as pd

# Função para obter dados da API
def get_pokemon_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        alerta(f"Erro ao acessar a API: {response.status_code}")
        return None

# Função de alerta
def alerta(mensagem):
    print(f"ALERTA: {mensagem}")

# URL da API
url = "https://pokeapi.co/api/v2/pokemon?limit=100"

# Obter dados dos Pokémon
data = get_pokemon_data(url)

# Listas para armazenar dados
pokemon_list = []
abilities_list = []
types_list = []

# Processar dados dos Pokémon
if data:
    for pokemon in data['results']:
        pokemon_data = get_pokemon_data(pokemon['url'])
        if pokemon_data:
            # Informações básicas
            pokemon_info = {
                'id': pokemon_data['id'],
                'name': pokemon_data['name'],
                'height': pokemon_data['height'],
                'weight': pokemon_data['weight']
            }
            pokemon_list.append(pokemon_info)
            
            # Habilidades
            for ability in pokemon_data['abilities']:
                abilities_list.append({
                    'pokemon_id': pokemon_data['id'],
                    'pokemon_name': pokemon_data['name'],
                    'ability_name': ability['ability']['name']
                })
            
            # Tipos
            for poke_type in pokemon_data['types']:
                types_list.append({
                    'pokemon_id': pokemon_data['id'],
                    'pokemon_name': pokemon_data['name'],
                    'type_name': poke_type['type']['name']
                })

# Criar DataFrames
pokemon_df = pd.DataFrame(pokemon_list)
abilities_df = pd.DataFrame(abilities_list)
types_df = pd.DataFrame(types_list)

# Exibir as tabelas
print("Tabela de Pokémon:")
print(pokemon_df.head())

print("\nTabela de Habilidades:")
print(abilities_df.head())

print("\nTabela de Tipos:")
print(types_df.head())
```
